```python
import matplotlib.pyplot as plt
   
Author = ["Longus", "Mark", "Mathew", "Ignatius"]
Cosine_Similarity = [0.097,0.404,0.103,0.997]

plt.barh(Author, Cosine_Similarity, color=['#004c6d', '#0083a6', '#00c0d8', '#00ffff'])
plt.title("Author vs. Cosine Similarity")
plt.xlabel('Cosine Similarity')
plt.ylabel('Author')
for index, value in enumerate(Cosine_Similarity):
    plt.text(value, index,
             str(value))
plt.savefig("CSTrain")
plt.show()
```
