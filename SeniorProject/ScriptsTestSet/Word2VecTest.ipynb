```python
import spacy
#processes file for W2V, tokenizes text if tokenize is true, then returns text
def process_file(a_file, tokenize=True):
    string_without_line_breaks = ""
    for line in a_file:
        stripped_line = line.rstrip() + " "
        string_without_line_breaks += stripped_line
    a_file.close()
    sentences = string_without_line_breaks
    
    
   
    if (tokenize == False):
        cleanedFile = cleanText(sentences, tokenize)
        return cleanedFile
    cleanedFile = cleanText(sentences) #cleanText is found in PreprocessLib.ipynb
        
    output_doc = []
    nlp = spacy.load("el_core_news_sm")

    for sentence in cleanedFile:
        temp = []
        doc = nlp(sentence)
        for token in doc:
            temp.append(token.text)
        output_doc.append(temp)
        output_doc = removeSingleLists(output_doc)
    return output_doc
```

```python
from nltk.corpus import PlaintextCorpusReader
corpus_root = './other-sources'
filelists = PlaintextCorpusReader(corpus_root, '.*')

#processes files for Word2Vec
longFile = open("./other-sources/Longus.txt", "r", encoding="utf8")
markFile = open("./other-sources/Mark.txt", "r", encoding="utf8")
mathFile = open("./other-sources/Mathew.txt", "r", encoding="utf8")
ignatFile = open("./other-sources/Ignatius-4.txt", "r", encoding="utf8")
testFile = open("./other-sources/Ignatius-3books.txt", "r", encoding="utf8")
long = process_file(longFile)
mark = process_file(markFile)
math = process_file(mathFile)
ignat = process_file(ignatFile)
test = process_file(testFile)
```

```python
import numpy as np
#computes cosine similarity between vector A and vector B, then returns result
def cosineSim(A, B):
    sim = np.dot(A, B) / (np.linalg.norm(A) * np.linalg.norm(B))
    if np.isnan(np.sum(sim)):
        return 0
    return sim
```

```python
import gensim
from gensim.models import Word2Vec
from sklearn.model_selection import ParameterGrid

parm_dict = {'min_count':(1,2,4),'alpha':(0.035,0.025,0.015), 'sample':(1.0E-4,1.0E-5,1.0E-6),
             'size':(5, 10, 20, 30, 100, 200), 'sg':(0,1)}

#this function tests two things
#1) the mean average cosine similarity of the similar words to the test document
#2) a count of which author's words are most similar to the test document's words
def runGrid():

    param_grid = ParameterGrid(parm_dict)
    for dict_ in param_grid:
        print(dict_)
        longModel = Word2Vec(min_count=dict_["min_count"], alpha=dict_["alpha"], sample=dict_["sample"],
                            size=dict_["size"], sg=dict_["sg"]) 
        longModel.build_vocab(long)
        longModel.train(long, total_examples=longModel.corpus_count, epochs=longModel.iter)

        markModel = Word2Vec(min_count=dict_["min_count"], alpha=dict_["alpha"], sample=dict_["sample"],
                            size=dict_["size"], sg=dict_["sg"]) 
        markModel.build_vocab(mark)
        markModel.train(mark, total_examples=markModel.corpus_count, epochs=markModel.iter)

        mathModel = Word2Vec(min_count=dict_["min_count"], alpha=dict_["alpha"], sample=dict_["sample"],
                            size=dict_["size"], sg=dict_["sg"]) 
        mathModel.build_vocab(math)
        mathModel.train(math, total_examples=mathModel.corpus_count, epochs=mathModel.iter)

        ignatModel = Word2Vec(min_count=dict_["min_count"], alpha=dict_["alpha"], sample=dict_["sample"],
                            size=dict_["size"], sg=dict_["sg"]) 
        ignatModel.build_vocab(ignat)
        ignatModel.train(ignat, total_examples=ignatModel.corpus_count, epochs=ignatModel.iter)

        testModel = Word2Vec(min_count=dict_["min_count"], alpha=dict_["alpha"], sample=dict_["sample"],
                            size=dict_["size"], sg=dict_["sg"]) 
        testModel.build_vocab(test)
        testModel.train(test, total_examples=testModel.corpus_count, epochs=testModel.iter)
        
        tDistList = []
        lDistList = []
        MrDistList = []
        MaDistList = []
        IgDistList = []
        
        countList = []

        for x in testModel.wv.vocab.keys():
            label = "notinany"
            maxNum = -9999

            count = 0

            if (x in testModel.wv.vocab):
                tDist = cosineSim(testModel[x], testModel[x])
                tDistList.append(tDist)
                
            if (x in longModel.wv.vocab):
                lDist = cosineSim(testModel[x], longModel[x])
                lDistList.append(lDist)
                count += 1
                if (lDist > maxNum):
                    maxNum = lDist
                    label = "Longus"

            if (x in markModel.wv.vocab):
                MrDist = cosineSim(testModel[x], markModel[x]) 
                MrDistList.append(MrDist)
                count += 1
                if (MrDist > maxNum):
                    maxNum = MrDist
                    label = "Mark"

            if (x in mathModel.wv.vocab):
                MaDist = cosineSim(testModel[x], mathModel[x])
                MaDistList.append(MaDist)
                count += 1
                if (MaDist > maxNum):
                    maxNum = MaDist
                    label = "Mathew"

            if (x in ignatModel.wv.vocab):   
                IgDist = cosineSim(testModel[x], ignatModel[x])
                IgDistList.append(IgDist)
                count += 1
                if (IgDist > maxNum):
                    maxNum = IgDist
                    label = "Ignatius"

            if (label != 'notinany' and count > 3):
                countList.append(label)

        import statistics
        print("Test:", statistics.mean(tDistList), "Longus:", statistics.mean(lDistList), "Mark:", statistics.mean(MrDistList), 
              "Math:", statistics.mean(MaDistList), "Ignatius:", statistics.mean(IgDistList))
        from collections import Counter
        from collections import OrderedDict
        list = Counter(countList)
        print(OrderedDict([(i, str(round(count / sum(list.values()) * 100.0, 3)) + '%') for i, count in list.most_common()]))
        
```

```python
import warnings
warnings.filterwarnings('ignore')
runGrid()

'''best results for simple average (count > 1)
{'alpha': 0.035, 'min_count': 2, 'sample': 0.0001, 'sg': 1, 'size': 200}
Test: 1.0 Longus: 0.09665793 Mark: 0.40384465 Math: 0.102080904 Ignatius: 0.9966392
count > 3:
{'alpha': 0.035, 'min_count': 2, 'sample': 0.0001, 'sg': 1, 'size': 200}
Test: 1.0 Longus: 0.096803255 Mark: 0.40384465 Math: 0.10250669 Ignatius: 0.9966392
'''

''' best results for word count (count > 1)
{'alpha': 0.025, 'min_count': 4, 'sample': 0.0001, 'sg': 1, 'size': 5}
OrderedDict([('Ignatius', '81.25%'), ('Mark', '18.75%')])
{'alpha': 0.025, 'min_count': 4, 'sample': 0.0001, 'sg': 1, 'size': 10}
OrderedDict([('Ignatius', '81.25%'), ('Mark', '18.75%')])
{'alpha': 0.025, 'min_count': 4, 'sample': 0.0001, 'sg': 1, 'size': 20}
OrderedDict([('Ignatius', '81.25%'), ('Mark', '18.75%')])
{'alpha': 0.025, 'min_count': 4, 'sample': 0.0001, 'sg': 1, 'size': 30}
OrderedDict([('Ignatius', '81.25%'), ('Mark', '18.75%')])
{'alpha': 0.025, 'min_count': 4, 'sample': 0.0001, 'sg': 1, 'size': 100}
OrderedDict([('Ignatius', '81.25%'), ('Mark', '18.75%')])
{'alpha': 0.025, 'min_count': 4, 'sample': 0.0001, 'sg': 1, 'size': 200}
count > 4:
{'alpha': 0.035, 'min_count': 1, 'sample': 0.0001, 'sg': 1, 'size': 200}
OrderedDict([('Ignatius', '100.0%')])
Note: there were many instances of this
'''
```
