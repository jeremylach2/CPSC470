```python
import spacy
#processes file for W2V, tokenizes text if tokenize is true, then returns text
def process_file(a_file, tokenize=True):
    string_without_line_breaks = ""
    for line in a_file:
        stripped_line = line.rstrip() + " "
        string_without_line_breaks += stripped_line
    a_file.close()
    sentences = string_without_line_breaks
    
    
   
    if (tokenize == False):
        cleanedFile = cleanText(sentences, tokenize)
        return cleanedFile
    cleanedFile = cleanText(sentences) #cleanText is found in PreprocessLib.ipynb
        
    output_doc = []
    nlp = spacy.load("el_core_news_sm")

    for sentence in cleanedFile:
        temp = []
        doc = nlp(sentence)
        for token in doc:
            temp.append(token.text)
        output_doc.append(temp)
        output_doc = removeSingleLists(output_doc)
    return output_doc
```

```python
#takes in a W2V model and returns the associated vocabulary of the model
def intoDict (model): 
    # Get the ordered list of words in the vocabulary
    words = model.wv.vocab.keys()
    # Make a dictionary
    we_dict = {word:model.wv[word] for word in words}
    return we_dict
```

```python
#takes a dictionary "dct", a list of words "wordList", and optionally a label
#takes each word vector of the words in wordList and adds them to an overall vector, then returns this vector
def createVec(dct, wordList, label="NA"):
    vec = []
    for word in wordList:
        vec.append(dct[word])
    if (label != "NA"):
        vec.append(label)
    return vec
```

```python
import gensim
import numpy as np
from gensim.models import Word2Vec
from sklearn.model_selection import ParameterGrid
from sklearn.neural_network import MLPClassifier

parm_dict = {'min_count':(1,2,4),'alpha':(0.035,0.025,0.015), 'sample':(1.0E-4,1.0E-5,1.0E-6),
             'size':(5, 10, 20, 30, 100, 200), 'sg':(0,1)}
#runs a grid search and prints the results of each iteration
def runGrid():
    param_grid = ParameterGrid(parm_dict)
    for dict_ in param_grid:
        print(dict_)
        import warnings
        warnings.filterwarnings('ignore')
        longModel = Word2Vec(min_count=dict_["min_count"], alpha=dict_["alpha"], sample=dict_["sample"],
                            size=dict_["size"], sg=dict_["sg"]) 
        longModel.build_vocab(long)
        longModel.train(long, total_examples=longModel.corpus_count, epochs=longModel.iter)

        markModel = Word2Vec(min_count=dict_["min_count"], alpha=dict_["alpha"], sample=dict_["sample"],
                            size=dict_["size"], sg=dict_["sg"]) 
        markModel.build_vocab(mark)
        markModel.train(mark, total_examples=markModel.corpus_count, epochs=markModel.iter)

        mathModel = Word2Vec(min_count=dict_["min_count"], alpha=dict_["alpha"], sample=dict_["sample"],
                            size=dict_["size"], sg=dict_["sg"]) 
        mathModel.build_vocab(math)
        mathModel.train(math, total_examples=mathModel.corpus_count, epochs=mathModel.iter)

        ignatModel = Word2Vec(min_count=dict_["min_count"], alpha=dict_["alpha"], sample=dict_["sample"],
                            size=dict_["size"], sg=dict_["sg"]) 
        ignatModel.build_vocab(ignat)
        ignatModel.train(ignat, total_examples=ignatModel.corpus_count, epochs=ignatModel.iter)

        testModel = Word2Vec(min_count=dict_["min_count"], alpha=dict_["alpha"], sample=dict_["sample"],
                            size=dict_["size"], sg=dict_["sg"]) 
        testModel.build_vocab(test)
        testModel.train(test, total_examples=testModel.corpus_count, epochs=testModel.iter)
        
        #creating feature vectors of the models
        longDict = intoDict(longModel)
        markDict = intoDict(markModel)
        mathDict = intoDict(mathModel)
        ignatDict = intoDict(ignatModel)
        testDict = intoDict(testModel)
        #creating vectors to represent the models based on the commonWordList
        commonWordList = ['πάντα', 'αὐτοῦ', 'ἐστιν', 'με', 'αὐτὸν', 'αὐτῶν', 'τοῦτο', 'ἵνα']
        longVec = createVec(longDict, commonWordList, 'Longus')
        markVec = createVec(markDict, commonWordList, 'Mark')
        mathVec = createVec(mathDict, commonWordList, 'Mathew')
        ignatVec = createVec(ignatDict, commonWordList, 'Ignatius')
        testVec = createVec(testDict, commonWordList)

        trainX = np.array((longVec[:5], markVec[:5], mathVec[:5], ignatVec[:5]))
        trainY =["Longus", "Mark", "Mathew", "Ignatius"]
        testX = np.array(testVec[:5])
        nsamples, nx, ny = trainX.shape
        trainXReshape = trainX.reshape((nsamples,nx*ny))

        clf = MLPClassifier(random_state=1, max_iter=2000, hidden_layer_sizes=(4,)).fit(trainXReshape, trainY)
        testXReshape = testX.reshape(1,testX.shape[1]*5)
        print(clf.predict_proba(testXReshape))
```

```python
#takes in a tag list and transforms it to 3D vector format
def prepareForW2V(tagList):
    cleaned = [[]]
    line = []
    for word in tagList:
        if (word[0] == "."):
            cleaned.append(line)
            line = []
        else:
            line.append(word[1])
    return cleaned
```

```python
#transforms the texts into vector format
longPrep = prepareForW2V(longTags)
markPrep = prepareForW2V(mathTags)
mathPrep = prepareForW2V(markTags)
ignatPrep = prepareForW2V(ignatTags)
testPrep = prepareForW2V(testTags)
```
