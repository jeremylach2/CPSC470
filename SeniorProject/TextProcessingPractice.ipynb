```python
from cltk import NLP
```

```python
cltk_nlp = NLP(language="grc")
```

```python
with open("grc-thucydides.txt", encoding="utf8") as fo: #make sure to specify utf-8 encoding
    grc_full = fo.read()
```

```python
grc_full
```

```python
print("Text snippet:", grc_full[:200])
print("Character count:", len(grc_full))
print("Approximate token count:", len(grc_full.split()))
```

```python
cltk_doc = cltk_nlp.analyze(text=grc_full)

# https://github.com/cltk/cltk/blob/master/notebooks/CLTK%20Demonstration.ipynb
```

```python
from cltk.tokenizers import GreekTokenizationProcess
from cltk.languages.example_texts import get_example_text
tokenizer_process = GreekTokenizationProcess()
output_doc = tokenizer_process.run(input_doc=cltk_doc)
output_doc.tokens
```
